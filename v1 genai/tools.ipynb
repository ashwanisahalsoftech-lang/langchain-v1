```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os
load_dotenv()
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
from langchain.tools import tool
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:qwen/qwen3-32b")
```

```python
@tool("personal_info")
def myinfo():
    """Information about myself(user/creator):my name is sayyam , 20 year old agentic ai learner ,i m pursuing btech from lj university , my height is 5'9 , skin complexion:fair"""
model_tool1=model.bind_tools([myinfo])
msg=[{"role":"user","content":"what is my name"}]
resp=model_tool1.invoke(msg)
msg.append(resp)
```

```python
for tool_call in resp.tool_calls:
    result=myinfo.invoke(tool_call)
    msg.append(result)

final_response=model_tool1.invoke(msg)
final_response.text
```
