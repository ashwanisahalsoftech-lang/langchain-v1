Streaming

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:groq/compound")
```

```python
resp=model.invoke("write a 500 words on essay")
print(resp.content)
```

```python
for chunk in model.stream("write a 500 essay words on cats"):
    print(chunk.text,end="",flush=True)
```

```python
resp=model.batch(["hi","what year is today","national currency of india"])
for re in resp:
    print(re.content)
```
